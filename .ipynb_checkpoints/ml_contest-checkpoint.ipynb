#### Исходные данные. Таблица 1 из 2. events_data

```python
events_data = pd.read_csv('event_data_train_1.csv', header=0).\
                        append(pd.read_csv('event_data_train_2.csv', header=0))
df = events_data
df[::(df.shape[0] // 50)]
```

*step_id* - id стэпа  
*user_id* - анонимизированный id юзера  
*timestamp* - время наступления события в формате unix date  
*action* - событие, возможные значения:  
+ *discovered* - пользователь перешел на стэп
+ *viewed* - просмотр шага,
+ *started_attempt* - начало попытки решить шаг, ранее нужно было явно нажать на кнопку - начать решение, перед тем как приступить к решению практического шага
+ *passed* - удачное решение практического шага


```python
%%script false
print(f'df.shape: {df.shape}')
for column in df:
    print('_' * 100)
    print(column, 
          df[column].dtypes, 
          f'notNaN fraction - {df[column].notna().sum() / df.shape[0]}', 
          f'notNaN - {df[column].notna().sum()}', 
          f'nunique - {df[column].nunique()}', 
          sep='\n')
    if df[column].dtype in ('object', 'bool'):
        print(
            pd.concat(
                [df[column].value_counts(), 
                 df[column].value_counts() / df[column].notna().sum()],  # why don't work parameter dropna=True?
                axis=1)
        )
        print('    Share among the 10 most common:')
        df[column].value_counts().head(10).plot(kind='pie')
        plt.show()
    elif df[column].dtype in ('float64', 'int64'):  # how write 'all digital types'?
        print(
            pd.concat(
                [df[column].value_counts(bins=10), 
                 df[column].value_counts(bins=10) / df[column].notna().sum()],  # why don't work parameter dropna=True?
                axis=1)
        )
        df[column].hist(bins=100, log=True)
        plt.show()
```

#### Создание дф Действия разных пользователей users_events_data

```python
# сколько каких действий совершил каждый пользователь
users_events_data = events_data.pivot_table(index='user_id', 
                                           columns='action', 
                                           values='step_id', 
                                           aggfunc='count', 
                                           fill_value=0) \
                                .reset_index()
users_events_data
```

#### Создание дф Количество (не)успешных действий пользователей users_scores

```python
# сколько (не)успешных сдач заданий совершил пользователь
users_scores = submissions_data.pivot_table(index='user_id', 
                                           columns='submission_status', 
                                           values='step_id', 
                                           aggfunc='count', 
                                           fill_value=0).reset_index()
users_scores
```

```python
%%script false
# гистограмма количество пользователей, успешно сдавших за время обучения x заданий
users_scores[users_scores.correct < 101].correct.hist(bins=100);
# users_scores.correct.max()  # 421 успешную сдачу кто-то сделал
# users_scores.correct[users_scores.correct > 100].count()  #  больше 100 успешных сдач у 18 человек
```

```python
%%script false
# гистограмма количество пользователей, не успешно сдавших за время обучения x заданий
users_scores[users_scores.wrong < 101].wrong.hist(bins=100);
```

#### Создание серии Перерывы в занятиях gap_data

```python
# довольно спорный подход из урока

# Берем только 3 столбца и выбираем только по одной строчке за день для каждого пользователя (верхнюю, вроде)
# группируем по пользователям и таймстемпы каждого пользователя
# складываем в список
# и вычисляем разницу между соседними таймстемпами пользователя (здесь выбросятся те, кто заходил один раз?)
# Берем только данные:  массив из кучи массивов (для каждого пользователя) с перерывами, которые делал юзер
gap_data = events_data[['user_id', 'day', 'timestamp']].drop_duplicates(subset=['user_id', 'day']) \
                                                        .groupby('user_id')['timestamp'] \
                                                        .apply(list) \
                                                        .apply(np.diff) \
                                                        .values

# сваливаем числа из подсписков в один единый список (теряя всю информацию об отдельных пользователях), 
# преобразуя данные в серию
gap_data = pd.Series(np.concatenate(gap_data, axis=0))

# переводим из секунд в сутки
gap_data = gap_data / (24 * 60 * 60)
gap_data
```

```python
%%script false
gap_data.hist(figsize=(12, 8), bins=100).set_yscale('log');
```

```python
%%script false
gap_data.quantile(0.90)
```

```python
%%script false
gap_data.quantile(0.95)
```

В 90% случаев ученики возвращаются в течение 18 дней, в 95% - 60 дней.   
Волюнтаристски взяли 30 дней как срок, по которому будем считать, что на курс ученик уже не вернется: drop_out_threshold

___  
#### Создание дф Данные по пользователям users_data

```python
users_data = events_data.groupby('user_id', as_index=False) \
                        .agg({'timestamp': 'max'}) \
                        .rename(columns={'timestamp': 'last_timestamp'})
users_data
```

```python
# время, когда выгрузили наши данные (последний таймстемп)
now = events_data.timestamp.max()
now
```

```python
# срок 30 дней в секундах
drop_out_threshold = 30 * 24 * 60 * 60
drop_out_threshold
```

```python
# добавляем колонку: является ли пользователь покинувшим курс 
users_data['is_gone_user'] = (now - users_data.last_timestamp) > drop_out_threshold
users_data
```

```python
users_data = users_data.merge(users_scores, on='user_id', how='outer')
users_data
```

```python
users_data = users_data.fillna(0)
users_data
```

```python
users_data = users_data.merge(users_events_data, how='outer')
users_data
```

```python
# сколько дней провел каждый юзер на курсе. Получается серия, которую преобразуем в дф
users_days = events_data.groupby('user_id').day.nunique().to_frame().reset_index()
users_days
```

```python
users_data = users_data.merge(users_days, how='outer')
users_data
```

```python
# проверяем не потеряли ли какие-то данные
users_data.user_id.nunique()
```

```python
# волюнтаристски полагаем прошедшими наш курс тех, кто решил более 170 заданий
users_data['passed_course'] = users_data.passed > 170
users_data[::(users_data.shape[0] // 50)]
```
